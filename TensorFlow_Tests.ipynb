```python
import tensorflow as tf

x = tf.Variable(3, name="x")
y = tf.Variable(4, name="y")
f = x*x*y + y + 2

init = tf.global_variables_initializer()
sess = tf.InteractiveSession()
init.run()
result = f.eval()
print(result)
sess.close()
```

```python
import numpy as np
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()
m, n = housing.data.shape
housing_data_plus_bias = np.c_[np.ones((m, 1)), housing.data]

X = tf.constant(housing_data_plus_bias, dtype=tf.float32, name="X")
y = tf.constant(housing.target.reshape(-1, 1), dtype=tf.float32, name="y")
XT = tf.transpose(X)
theta = tf.matmul(tf.matmul(tf.matrix_inverse(tf.matmul(XT, X)), XT), y)

with tf.Session() as sess:
    theta_value = theta.eval()
```

```python
import os
import pandas as pd

MNIST_PATH = os.path.join("datasets", "mnist")

def load_mnist_dataset():
    csv_path = os.path.join(MNIST_PATH, "kaggle_mnist_train.csv")
    return pd.read_csv(csv_path)
```

```python
mnist_data = load_mnist_dataset()
y_train = mnist_data["label"].as_matrix()
X_train = mnist_data.drop("label", axis=1).as_matrix()

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train.astype(np.float64))
```

```python
import tensorflow as tf

feature_cols = tf.contrib.learn.infer_real_valued_columns_from_input(X_train)
dnn_clf = tf.contrib.learn.DNNClassifier(hidden_units=[300, 100], n_classes=10, feature_columns=feature_cols)
dnn_clf = tf.contrib.learn.SKCompat(dnn_clf)

dnn_clf.fit(X_train_scaled, y_train, batch_size=50, steps=40000)
```

```python
import pandas as pd
import numpy as np

def load_mnist_test_dataset():
    csv_path = os.path.join(MNIST_PATH, "kaggle_mnist_test.csv")
    return pd.read_csv(csv_path)

mnist_test = load_mnist_test_dataset()
X_test = mnist_test.as_matrix()
predictions = dnn_clf.predict(X_test)
```

```python
df = pd.DataFrame(data = {'ImageId':np.arange(1, len(X_test) + 1, 1), 'Label':predictions['classes']})
df.to_csv('test_dnn.csv', index=False)
```
